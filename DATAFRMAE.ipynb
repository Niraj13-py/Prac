# Types of Joins in SQL

## 1. Inner Join
# Only records with matching values in both tables are included.

## 2. Left Join (Left Outer Join)
# Includes **all records** from the left table and only the **matching records** from the right table.

## 3. Right Join (Right Outer Join)
# Includes **all records** from the right table and only the **matching records** from the left table.

## 4. Full Outer Join
# Includes **all records** from both tables, with matches where available. Unmatched rows are filled with NULLs.


```python
#Inner join
df_inner = pd.merge(left=df_students, right=df_depts, left_on='Dept', right_on='dept_id', how='inner')
df_inner
```
