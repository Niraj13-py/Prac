```python
np.random.seed(50)
marks = np.random.randint(0, 100, 20)
marks
```

```python
# To give grace marks to students who scored less than 50
grace_marks = np.where(marks < 50, marks + 10, marks)    
print(grace_marks)
```

```python
grace_marks[marks < 50]
```

```python
import numpy as np
np.random.seed(1)
marks = np.random.randint(0, 100, (10,4))
print(marks)
```

```python
avg_marks = np.mean(marks, axis=1)
print(avg_marks)
```

```python
#average marks of all students 
avg_marks = np.mean(marks, axis=0)
print(avg_marks)    
```
