```python
import pandas as pd
A = {'Product_Brand': ['Apple', 'Apple', 'Samsung', 'Samsung', 'Samsung', 'LG', 'LG'], 
'Product_Name': ['Mobile', 'Laptop', 'Mobile', 'Laptop', 'LCD', 'LCD', 'Laptop'], 
'Price': [70000, 150000, 140000,60000, 20000,15000,30000],
'Rating': [4.9,4.4,4.8,3.9,4.1,3.7,3.6]}
```

```python
#delete price column
df = pd.DataFrame(A, index = ['A', 'B', 'C', 'D', 'E', 'F', 'G'])   
df.drop('Price', axis=1, inplace=True)  
```

```python
#Delete rows 
df.drop(['A', 'B'], axis=0, inplace=True)
```

```python
df
```

```python
#delete samsung row     
df.drop(df[df['Product_Brand'] == 'Samsung'].index, inplace=True)
```
