```python
def get_Final_Price(brand_name, price):  
    if brand_name == 'Apple':
        return price - (price * 0.2)  # 20% discount for Apple
    elif brand_name == 'Samsung':
        return price - (price * 0.3)  # 15% discount for Samsung
    elif brand_name == 'LG':
        return price - (price * 0.05)  # 5% discount for LG
    else:
        return price  
```

```python
def get_Final_Price(row):
    brand_name = row['Product_Brand']
    price = row['Price']
    if brand_name.lower() == 'Apple':
        disc = 20
    elif brand_name.lower() == 'Samsung':
        disc = 15
    else:
        disc = 10       
        return price - (price * disc / 100)
```

```python

df['Final Price'] = df.apply(get_Final_Price, axis=1)  #Apply the function to each row of the DataFrame
```

```python
#rating based on price
def get_Rating(row):
    price = row['Price']
    if price > 100000:
        return 5.0
    elif price > 50000:
        return 4.5
    else:
        return 3.5
```
