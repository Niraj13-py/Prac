```python
# Python One-Shot Reference Guide

# ============= BASICS =============
# Variables and Types
name = "Alice"
age = 30
height = 5.7
is_student = True
numbers = [1, 2, 3, 4, 5]
person = {"name": "Bob", "age": 25}

# String Operations
text = "Hello World"
print(f"Length: {len(text)}")
print(f"Upper: {text.upper()}")
print(f"Split: {text.split()}")
print(f"Replace: {text.replace('World', 'Python')}")
print(f"Format: {name} is {age} years old")

# ============= CONTROL FLOW =============
# Conditionals
if age >= 18:
    print("Adult")
elif age >= 13:
    print("Teen")
else:
    print("Child")

# Loops
for i in range(5):
    print(f"Count: {i}")

for item in numbers:
    print(f"Number: {item}")

while age > 0:
    age -= 1
    if age == 25:
        break

# ============= DATA STRUCTURES =============
# Lists
fruits = ["apple", "banana", "orange"]
fruits.append("grape")
fruits.remove("banana")
print(f"First fruit: {fruits[0]}")
print(f"Last fruit: {fruits[-1]}")

# List Comprehension
squares = [x**2 for x in range(10)]
evens = [x for x in range(20) if x % 2 == 0]

# Dictionaries
student = {
    "name": "Charlie",
    "grade": 85,
    "subjects": ["math", "science"]
}
print(f"Name: {student['name']}")
print(f"Grade: {student.get('grade', 'N/A')}")

# Sets
unique_numbers = {1, 2, 3, 3, 4, 5}  # Duplicates removed
print(f"Unique: {unique_numbers}")

# Tuples (immutable)
coordinates = (10, 20)
x, y = coordinates  # Unpacking

# ============= FUNCTIONS =============
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

def calculate_area(length, width):
    return length * width

# Lambda functions
add = lambda x, y: x + y
print(f"Sum: {add(5, 3)}")

# Function with *args and **kwargs
def flexible_function(*args, **kwargs):
    print(f"Args: {args}")
    print(f"Kwargs: {kwargs}")

flexible_function(1, 2, 3, name="Alice", age=30)

# ============= CLASSES =============
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def introduce(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old"
    
    def have_birthday(self):
        self.age += 1

# Create and use object
person = Person("David", 28)
print(person.introduce())
person.have_birthday()
print(f"After birthday: {person.age}")

# ============= FILE OPERATIONS =============
# Writing to file
with open("example.txt", "w") as f:
    f.write("Hello, Python!\n")
    f.write("This is a test file.")

# Reading from file
try:
    with open("example.txt", "r") as f:
        content = f.read()
        print(f"File content: {content}")
except FileNotFoundError:
    print("File not found")

# ============= ERROR HANDLING =============
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")
finally:
    print("Cleanup code here")

# ============= USEFUL BUILT-INS =============
# Map, Filter, Reduce
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
evens = list(filter(lambda x: x % 2 == 0, numbers))

from functools import reduce
sum_all = reduce(lambda x, y: x + y, numbers)

# Enumerate and Zip
for index, value in enumerate(fruits):
    print(f"Index {index}: {value}")

names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(f"{name} is {age} years old")

# ============= COMMON PATTERNS =============
# Check if all/any items meet condition
numbers = [2, 4, 6, 8]
print(f"All even: {all(x % 2 == 0 for x in numbers)}")
print(f"Any > 5: {any(x > 5 for x in numbers)}")

# Sorting
students = [
    {"name": "Alice", "grade": 85},
    {"name": "Bob", "grade": 92},
    {"name": "Charlie", "grade": 78}
]
sorted_students = sorted(students, key=lambda x: x["grade"], reverse=True)

# Dictionary comprehension
word_lengths = {word: len(word) for word in ["hello", "world", "python"]}

# ============= MODULES AND IMPORTS =============
import math
from datetime import datetime
import random as rand

print(f"Pi: {math.pi}")
print(f"Current time: {datetime.now()}")
print(f"Random number: {rand.randint(1, 100)}")

# ============= USEFUL ONE-LINERS =============
# Flatten nested list
nested = [[1, 2], [3, 4], [5, 6]]
flattened = [item for sublist in nested for item in sublist]

# Count occurrences
text = "hello world"
char_count = {char: text.count(char) for char in set(text)}

# Remove duplicates while preserving order
items = [1, 2, 2, 3, 1, 4]
unique_ordered = list(dict.fromkeys(items))

# Swap variables
a, b = 10, 20
a, b = b, a

# Multiple assignment
name, age, city = "Alice", 30, "New York"

# ============= DEBUGGING =============
# Print with variable names
x = 10
y = 20
print(f"{x=}, {y=}")  # Prints: x=10, y=20

# Quick type checking
print(f"Type of x: {type(x)}")
print(f"Is x an int? {isinstance(x, int)}")

# ============= PERFORMANCE TIPS =============
# Use list comprehensions instead of loops when possible
# Use sets for membership testing (O(1) vs O(n))
# Use enumerate() instead of range(len())
# Use zip() for parallel iteration
# Use dict.get() instead of key checking

print("Python one-shot reference complete!")
```
